```python
import warnings

import pandas as pd

from jre_utils.datapath import DATA_DIRECTORY_PATH

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", None)
```

```python
taxable_income_with_area_code_path = (
    f"{DATA_DIRECTORY_PATH}/exploratory/taxable_income_with_area_code.csv"
)
taxable_income_with_area_code_df = pd.read_csv(taxable_income_with_area_code_path)
taxable_income_with_area_code_df = taxable_income_with_area_code_df[taxable_income_with_area_code_df["year"] == 2021]
taxable_income_with_area_code_df
```

```python
area_code_to_area = pd.Series(
    taxable_income_with_area_code_df["area"],
    index=taxable_income_with_area_code_df["area_code"],
).to_dict()
```

```python
local_finance_survey_cover = f"{DATA_DIRECTORY_PATH}/exploratory/local_finance_survey_cover.csv"
local_finance_survey_cover_df = pd.read_csv(local_finance_survey_cover)
local_finance_survey_cover_df = local_finance_survey_cover_df[local_finance_survey_cover_df["year"] == 2021]
local_finance_survey_cover_df = local_finance_survey_cover_df[
    local_finance_survey_cover_df["area_code"].apply(lambda x: x.isdigit())
]
local_finance_survey_cover_df["area_code"] = local_finance_survey_cover_df["area_code"].astype(int)
local_finance_survey_cover_df["area_code_shortened"] = local_finance_survey_cover_df["area_code"] // 10
local_finance_survey_cover_df
```

```python
taxable_income_with_area_code_df.merge(local_finance_survey_cover_df, left_on="area_code", right_on="area_code_shortened")
```
